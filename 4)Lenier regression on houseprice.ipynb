```python
drop_values=['Id','Alley','MiscFeature','PoolQC']
df.drop(drop_values,axis=1,inplace=True)         #When inplace = True , the data is modified in place
df.head()
```

```python
df.info()   #df.info() provides info of index dtype and column dtypes, non-null values and memory usage.
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].median(),inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].median(),inplace=True)
df['GarageYrBlt'].fillna(df['GarageYrBlt'].median(),inplace=True)
```

```python
C_feat=df.select_dtypes(include = "object").columns

from sklearn.preprocessing import LabelEncoder     
le=LabelEncoder()
for feature in C_feat:
  df[feature]=le.fit_transform(df[feature].astype(str))
df.info()
```

```python
C_feat
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['SalePrice'], axis = 1)
y = df.loc[:,['SalePrice']].values
sc=StandardScaler()
X=sc.fit_transform(X)
y=y.reshape(-1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
clf=LinearRegression()
reg=clf.fit(X_train, y_train)
predictions = reg.predict(X_test)
print('MSE : ',mean_squared_error(y_test,predictions))
print('R2_score:',r2_score(y_test,predictions))
```

```python
output=pd.DataFrame([y_test,predictions],index=['REAL' ,'PREDICTED'])
output.transpose()
```

Ridge Regression is a technique for analyzing multiple regression data that suffer from multicollinearity.
Grid-searching will build a model on each parameter combination possible.
Repeated k-fold cross-validation provides a way to improve the estimated performance of a machine learning model.
n_splitsint. Returns the number of splitting iterations in the cross-validator.


```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
cv=RepeatedKFold(n_splits=20,n_repeats=5,random_state=1)
grid=dict()
grid['alpha']=np.arange(0,1,0.1)
ridge=Ridge()
ridge_reg=GridSearchCV(ridge,grid,scoring='neg_mean_absolute_error',cv=cv,n_jobs=-1)
ridge_reg.fit(X,y)
```

```python
ridge_pred=ridge_reg.predict(X_test)
print('r2_score of RIDGE REGREESOR :',r2_score(y_test,ridge_pred)*100)
```
